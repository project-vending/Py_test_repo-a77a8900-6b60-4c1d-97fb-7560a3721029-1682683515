```
# Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```
# Reading the csv file into a pandas dataframe
df = pd.read_csv("https://github.com/midhunmohank/INFO6105/blob/main/advertising.csv")

```

```
# Getting a summary of the data
df.head()

```

```
# Checking for null values
df.isnull().sum()

```

```
# Visualizing the distribution of the data using histograms and KDE plots
sns.displot(df['A'], kde=True)
plt.title("A Distribution")
plt.xlabel("A")
plt.show()

sns.displot(df['B'], kde=True)
plt.title("B Distribution")
plt.xlabel("B")
plt.show()

sns.displot(df['C'], kde=True)
plt.title("C Distribution")
plt.xlabel("C")
plt.show() 

```

```
# Visualizing the relationship between the variables using scatter plots
sns.scatterplot(x="A", y="B", data=df)
plt.title("A vs B")
plt.show()

sns.scatterplot(x="B", y="C", data=df)
plt.title("B vs C")
plt.show()

sns.scatterplot(x="C", y="A", data=df)
plt.title("C vs A")
plt.show()

```

```
# Checking for correlation between the variables
sns.heatmap(df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
